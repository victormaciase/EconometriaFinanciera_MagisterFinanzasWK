# ¿Qué es una librería en Python?

Una librería (o módulo) es una colección de funciones, herramientas y códigos preescritos que extienden las capacidades básicas de Python. En lugar de programar todo desde cero para resolver un problema, puedes usar una librería para ahorrar tiempo y esfuerzo.

Existen dos tipos principales:

- **Estándar:** Vienen incluidas por defecto en Python (como *math* o *random*).

- **Externas:** Desarrolladas por la comunidad (como *pandas* para analizar datos o *matplotlib* para hacer gráficos) que se instalan fácilmente.

# ¿Cómo se usan?
Para usar una librería en tu código, solo debes importarla al inicio de tu archivo usando la palabra clave *import*.

```python
# 1. Importas la librería (puedes darle un alias corto con 'as')
import numpy as np

# 2. Usas sus funciones predefinidas
datos = [4, 2, 1, 9, 8]
promedio = np.mean(datos)

print(promedio) # Devuelve 4.8 sin que hayas tenido que programar la fórmula matemática
```
